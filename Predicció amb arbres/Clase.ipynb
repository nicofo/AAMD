### Regresión con un predictor continuo

```r
library(rpart)
x<-seq(-3,3,by=0.1)
y<-x^2
plot(x,y, "l",lwd=2,col="blue")
```

```r
r<-rpart(y~x)
plot(r,branch=0.5)
text(r,use.n=TRUE,xpd=2)
```

```r
yp<-predict(r,newdata=data.frame(x))
plot(x,y, "l",lwd=2,col="blue")
lines(x,yp,lwd=2,col="red")
```

### Clasificación con rpart

```r
library(MASS)
data(synth.tr)
data(synth.te)
str(synth.tr)
```

```r
synth.tr$yc<-factor(synth.tr$yc)
synth.te$yc<-factor(synth.te$yc)
```

```r
srp<-rpart(yc~xs+ys,data=synth.tr)
```

```r
plot(srp)
text(srp,use.n=TRUE,xpd=2)
```

```r
ycp <- predict(srp,synth.te, type = "class")
C<-table(True = synth.te$yc, Pred = ycp)
C
error.Clas<-1-sum(diag(C)/sum(C))
```

```r
library(ElemStatLearn)
data(spam)
```

```r
ITest<-sample(1:nrow(spam),nrow(spam)/10,replace=FALSE)
spam.train<-spam[-ITest,]
spam.test<-spam[ITest,]
srp<-rpart(spam~.,data=spam.train)
```

```r
plot(srp)
text(srp,use.n=TRUE,xpd=2)
```

```r
ycp <- predict(srp,spam.test, type = "class")
C<-table(True = spam.test$spam, Pred = ycp)
C
error.Clas<-1-sum(diag(C)/sum(C))
error.Clas
```
