```r
library(ISLR)
data(Auto)
help(Auto)
```

Eliminem columna name, per que hi han pocs repetits, i eliminem origin, ja que no ha de ser una bona predictora.

```r
Auto<-Auto[,-9]
Auto<-Auto[,-8]
```

#### Analitzem les variables
Ara analizarem cada variable del dataset.
##### mpg
Consum del vehicle en milles per galó de combustible.

Aquestes dades presenten una distibució més o meny normal.


```r
summary(Auto$mpg)
boxplot(Auto$mpg)
hist(Auto$mpg)
```

##### cylinders
Nombre de cilindres(entre 4 i 8).

Es veu una gran concentració de 4 cilindres, després de 8, tot següit 6 i posteriorment 5 i 3.



```r
summary(Auto$cylinders)
boxplot(Auto$cylinders)
hist(Auto$cylinders)
```

##### displacement
Desplaçament del motor.

No te una distibució normal. Els valors flucutan molt.

```r
summary(Auto$displacement)
boxplot(Auto$displacement)
hist(Auto$displacement)
```

##### horsepower
Cavalls del veichle.

Aquestes grafiques ja semblen mes normalitzades peró encara estan molt tirant cap a la dreta.


```r
summary(Auto$horsepower)
boxplot(Auto$horsepower)
hist(Auto$horsepower)
```

##### weight
Pes del vehicle



```r
summary(Auto$weight)
boxplot(Auto$weight)
hist(Auto$weight)
```

##### acceleration
Aceleració en milles per hora

Tant la mitjana com la mdeiana estan en el mateix valor. Se veu uan distibució normal.

```r
summary(Auto$acceleration)
boxplot(Auto$acceleration)
hist(Auto$acceleration)
```

##### year
Any del model.

No descriuen cap distibució normal.

```r
summary(Auto$year)
boxplot(Auto$year)
hist(Auto$year)
```

Anem a comprobar quin son els millors predictors. 

Primer utilitzarem el metode forward.

```r
ls4<-lm(mpg~1, data=Auto)
step(ls4,"~cylinders+displacement+horsepower+weight+acceleration+year")
```

Ara el metode backward

```r
ls5<-lm(mpg~cylinders+displacement+horsepower+weight+acceleration+year, data=Auto)
step(ls5)
```

Amb tot dos metodes em vist que la millor formula és **mpg ~ weight + year**
